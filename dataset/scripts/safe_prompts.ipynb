# Safe prompts

We used AI generated prompts (500 safe + 100 safe with tricky words)

We specifically chose **command** type of prompts as our unsafe prompt dataset also contains this kind prompts.

```python
import pandas as pd

df1 = pd.read_csv('/content/drive/MyDrive/.../safe.csv') # This path needs to be updated
df2 = pd.read_csv('/content/drive/MyDrive/.../tricky_safe.csv') # This path needs to be updated
```

```python
df1.head()
```

```python
df2.head()
```

```python
# added safeness type column to check later its performance

df1['safeness type'] = 'safe'
df2['safeness type'] = 'tricky safe'
```

```python
# keep important columns

columns = ['Prompt', 'Prompt_Type', 'harmful', 'safeness type']
combined = pd.concat([df1[columns], df2[columns]], ignore_index = True)
```

```python
# shuffle the rows
combined = combined.sample(frac=1, random_state=42).reset_index(drop=True)
combined.insert(0, 'id', range(1,len(combined)+1))
```
