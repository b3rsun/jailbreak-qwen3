# Mixed prompts
In this notebook, we mixed 600 safe + 520 unsafe prompts.


```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
import pandas as pd

safe = pd.read_csv('/content/drive/MyDrive/.../safe_prompts.csv') # This path needs to be updated
unsafe = pd.read_csv('/content/drive/MyDrive/.../harmful_behaviors.csv') # This path needs to be updated

print(safe.columns.tolist())
print(unsafe.columns.tolist())
```

```python
# change the advbench data's columns

unsafe = unsafe.rename(columns={'goal': 'Prompt'})
unsafe['harmful'] = 1
unsafe['type'] = 'unsafe'
unsafe['Prompt_Type'] = 'Command'
```

```python
# keep shared columns
columns = ['Prompt', 'Prompt_Type', 'harmful', 'type']
safe = safe.rename(columns={'safeness type': 'type'})
safe = safe[columns].copy()
unsafe = unsafe[columns].copy()
```

```python
# merge and shuffle

combined = pd.concat([safe, unsafe], ignore_index=True)
combined = combined.sample(frac=1, random_state=42).reset_index(drop=True)
combined.insert(0, 'id', range(1, len(combined) + 1))
```

```python
print(combined.shape)
print(combined['harmful'].value_counts())
print(combined['type'].value_counts())
combined.head(10)
```

```python
combined.to_csv('/content/drive/MyDrive/.../mixed_prompts.csv', index=False) # This path needs to be updated
```
